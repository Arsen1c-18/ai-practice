```python
#practice:
# water jug
from math import gcd

def water_jug_manual():
    cap1 = int(input("Enter capacity of Jug1: "))
    cap2 = int(input("Enter capacity of Jug2: "))
    target = int(input("Enter target: "))

    # Feasibility check
    if target > max(cap1, cap2) or target % gcd(cap1, cap2) != 0:
        print("No solution possible")
        return

    jug1, jug2 = 0, 0

    while True:
        print(f"\nJug1:{jug1}, Jug2:{jug2}")

        if jug1 == target or jug2 == target:
            print("Final state reached")
            break

        print("\n1.Fill Jug1\n2.Fill Jug2\n3.Empty Jug1\n4.Empty Jug2")
        print("5.Pour Jug1→Jug2\n6.Pour Jug2→Jug1\n0.Exit")

        move = int(input("Enter move: "))

        if move == 1: jug1 = cap1
        elif move == 2: jug2 = cap2
        elif move == 3: jug1 = 0
        elif move == 4: jug2 = 0

        elif move == 5:
            pour = min(jug1, cap2 - jug2)
            jug1 -= pour
            jug2 += pour

        elif move == 6:
            pour = min(jug2, cap1 - jug1)
            jug2 -= pour
            jug1 += pour

        elif move == 0:
            break
        else:
            print("Invalid move")

water_jug_manual()
```

```python
#water jub auto
def water_jug_auto(capA, capB, target):
    visited = set()
    queue = [((0,0), [(0,0)])]

    while queue:
        (a,b), path = queue.pop(0)

        if (a,b) in visited:
            continue
        visited.add((a,b))

        if a == target or b == target:
            print("Solution Path:")
            for state in path:
                print(state)
            return

        next_states = [
            (capA, b), (a, capB),
            (0, b), (a, 0)
        ]

        pour = min(a, capB-b)
        next_states.append((a-pour, b+pour))

        pour = min(b, capA-a)
        next_states.append((a+pour, b-pour))

        for state in next_states:
            if state not in visited:
                queue.append((state, path + [state]))

    print("No solution found")

water_jug_auto(4,3,2)
```

```python
#bfs
from collections import deque

grid = [
[0,1,0,0],
[0,0,0,1],
[1,0,1,0],
[0,0,0,0]
]

def is_valid(r,c,grid,visited):
    rows, cols = len(grid), len(grid[0])
    if not (0<=r<rows and 0<=c<cols):
        return False
    if grid[r][c]==1 or (r,c) in visited:
        return False
    return True

def bfs_grid_path(grid,start,goal):
    moves=[(0,1),(0,-1),(1,0),(-1,0)]
    queue=deque([(start,[start])])
    visited={start}

    while queue:
        (r,c),path=queue.popleft()

        if (r,c)==goal:
            return path

        for dr,dc in moves:
            nr,nc=r+dr,c+dc
            if is_valid(nr,nc,grid,visited):
                visited.add((nr,nc))
                queue.append(((nr,nc),path+[(nr,nc)]))

    return "No path found"

print(bfs_grid_path(grid,(0,0),(3,3)))
```

```python
#dls
graph = {
'A':['B','C','D'],
'B':['E','F'],
'C':['G'],
'D':['H'],
'E':['I'],
'F':['J','K'],
'G':['L'],
'H':['M','N'],
'I':[], 'J':[], 'K':['O','P'],
'L':['R'], 'M':[], 'N':['S'],
'O':[], 'P':[], 'R':[], 'S':[]
}

def dls(node, limit, path):
    if limit < 0:
        return None

    path.append(node)

    if limit == 0:
        return path.copy()

    for child in graph.get(node,[]):
        result = dls(child, limit-1, path)
        if result:
            return result

    path.pop()
    return None

print(dls('A',2,[]))
```

```python
#iddls
def dls(node, goal, limit, path):
    if limit < 0:
        return None

    path.append(node)

    if node == goal:
        return path.copy()

    for child in graph.get(node,[]):
        result = dls(child, goal, limit-1, path)
        if result:
            return result

    path.pop()
    return None

def iddfs(start, goal, max_depth=10):
    for depth in range(max_depth+1):
        path = dls(start, goal, depth, [])
        if path:
            return path
    return None

print(iddfs('A','R'))
```

```python
#ucs
import heapq

graph = {
"Source":[("A",1),("B",3),("C",5)],
"A":[("G",3)],
"B":[("F",5)],
"C":[("E",2)],
"E":[("Dest",1)],
"G":[], "F":[], "Dest":[]
}

def uniform_cost_search(start,goal):
    pq=[(0,start,[start])]
    visited=set()

    while pq:
        cost,node,path=heapq.heappop(pq)

        if node==goal:
            return path,cost

        if node in visited:
            continue
        visited.add(node)

        for neighbor,w in graph[node]:
            heapq.heappush(pq,(cost+w,neighbor,path+[neighbor]))

    return None,float("inf")

print(uniform_cost_search("Source","Dest"))
```
